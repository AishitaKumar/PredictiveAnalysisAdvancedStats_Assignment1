```python
import numpy as np
import pandas as pd
import math
```

```python
roll = 102303744

a_r = 0.05 * (roll % 7)        
b_r = 0.3 * ((roll % 5) + 1) 
```

```python
df = pd.read_csv("data.csv", encoding="latin1") 
x = pd.to_numeric(df["no2"], errors="coerce").dropna().to_numpy()
```

```python
z = x + a_r * np.sin(b_r * x)

mu = np.mean(z)
sigma2 = np.mean((z - mu) ** 2)    

lam = 1 / (2 * sigma2)
c = 1 / math.sqrt(2 * math.pi * sigma2)
```

```python
print("Roll number:", roll)
print("a_r =", a_r)
print("b_r =", b_r)
print("mu =", mu)
print("lambda =", lam)
print("c =", c)

print("Check c = sqrt(lambda/pi):", math.sqrt(lam / math.pi))
```
